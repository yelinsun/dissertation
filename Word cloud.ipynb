```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Sample text
text = """
Where the need for outdoor space is greatest, small vacant lots and land occupied by derelict buildings can be acquired, usually at low cost, to be developed into small parks
Parks smaller than most urban parks; less than half an acre in size; and are often created out of vacant lots, rooftops and otherwise forgotten and unused spaces
Small parks are a key part of most neighborhoods, but they typically provide mostly recreational benefits. They have limited areas, so they cannot meet all the potential demands for space for varied human activities and multiple natural processes
Pocket parks are urban open spaces at the very small scale. Usually only a few house lots in size or smaller, pocket parks can be tucked into and scattered throughout the urban fabric where they serve the immediately local population
The concept of a pocket park emerged from the need to improve the quality of urban life in the aspect of accessibility of public spaces, including green areas
Pocket parks are generally recognized as public park spaces that occupy less than one acre of land, the shape and dimensions of a pocket park will vary depending on the site
Pocket parks are well-designed small urban spaces with, or without, green elements, for urban quality, and urban resilience, wellbeing, connectivity, diversity and flexibility of use
Pocket parks can be defined as low-cost, small green open spaces for various activities and benefits
Pocket parks are tiny urban parks created from vacant lots that are accessible to the public. They can be in the urban fabric of different areas and are typically smaller than 1–3 acres or between 1 and 4 house lots in size
Pocket parks are defined as small green spaces embedded in, but not separate from, the urban landscape which offer small spaces of solace amidst the busy city streets with profound impact
Pocket parks are parks formed by further adding leisure and entertainment facilities on the basis of planting greenery on smaller plots. Pocket park construction areas are generally one to three times the area of housing construction
A park greening activity site open to the public, with a small scale, diverse shapes, and certain recreational functions, generally covering an area of between 400 and 10 000 square meters
The pocket parks are usually small green urban spaces that are flexibly distributed in high-density urban areas as patches
"""

# Setting up stopwords, you can add more to this set if you wish
stopwords = set(STOPWORDS)

# Create the WordCloud object
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=800, height=600).generate(text)

# Plot the WordCloud
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
import pandas as pd
from collections import Counter
import re
from nltk.corpus import stopwords

# Sample text
text = """
Where the need for outdoor space is greatest, small vacant lots and land occupied by derelict buildings can be acquired, usually at low cost, to be developed into small parks
Parks smaller than most urban parks; less than half an acre in size; and are often created out of vacant lots, rooftops and otherwise forgotten and unused spaces
Small parks are a key part of most neighborhoods, but they typically provide mostly recreational benefits. They have limited areas, so they cannot meet all the potential demands for space for varied human activities and multiple natural processes
Pocket parks are urban open spaces at the very small scale. Usually only a few house lots in size or smaller, pocket parks can be tucked into and scattered throughout the urban fabric where they serve the immediately local population
The concept of a pocket park emerged from the need to improve the quality of urban life in the aspect of accessibility of public spaces, including green areas
Pocket parks are generally recognized as public park spaces that occupy less than one acre of land, the shape and dimensions of a pocket park will vary depending on the site
Pocket parks are well-designed small urban spaces with, or without, green elements, for urban quality, and urban resilience, wellbeing, connectivity, diversity and flexibility of use
Pocket parks can be defined as low-cost, small green open spaces for various activities and benefits
Pocket parks are tiny urban parks created from vacant lots that are accessible to the public. They can be in the urban fabric of different areas and are typically smaller than 1–3 acres or between 1 and 4 house lots in size
Pocket parks are defined as small green spaces embedded in, but not separate from, the urban landscape which offer small spaces of solace amidst the busy city streets with profound impact
Pocket parks are parks formed by further adding leisure and entertainment facilities on the basis of planting greenery on smaller plots. Pocket park construction areas are generally one to three times the area of housing construction
A park greening activity site open to the public, with a small scale, diverse shapes, and certain recreational functions, generally covering an area of between 400 and 10 000 square meters
The pocket parks are usually small green urban spaces that are flexibly distributed in high-density urban areas as patches
"""

# Getting English stopwords from NLTK
stop_words = set(stopwords.words('english'))

# Cleaning the text
text = re.sub(r'[^A-Za-z\s]', '', text).lower()  # Remove non-alphabetical characters and convert to lowercase

# Split the text into words and filter out stopwords
words = [word for word in text.split() if word not in stop_words]

# Count word frequencies
word_frequencies = Counter(words)

# Convert the Counter object into a pandas DataFrame
df = pd.DataFrame(word_frequencies.items(), columns=['Word', 'Frequency'])

# Sort the DataFrame by frequency
df = df.sort_values(by='Frequency', ascending=False)

# Display the table
print(df)
```
